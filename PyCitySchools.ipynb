```python
# Dependencies and Setup

import pandas as pd

# File to Load (Remember to Change These)

school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
#school_data.head()
#school_data.shape()
#school_data.head()
#student_data.shape()
# Combine the data into a single dataset

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.shape
#school_data_complete.head()

school_data_complete
```

```python
#Calculate the total number of schools:

total_number_schools = len(school_data_complete["School ID"].unique())
total_number_schools
```

```python
#Calculate the total number of students:

total_number_students = len(school_data_complete["Student ID"].unique())
total_number_students
```

```python
#Calculate the total budget:

total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calculate the average math score

avg_math_score = student_data["math_score"].mean()
avg_math_score
```

```python
#Calculate the average reading score

avg_reading_score = student_data["reading_score"].mean()
avg_reading_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

num_pass_math = sum(student_data["math_score"] >= 70)
perc_pass_math = (num_pass_math/total_number_students) * 100
perc_pass_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)

num_pass_reading = sum(student_data["reading_score"] >= 70)
perc_pass_reading = (num_pass_reading/total_number_students) * 100
perc_pass_reading
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)

num_pass_both = sum((student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70))
overall_percent = num_pass_both/total_number_students * 100
overall_percent
```

```python
#District Summary Table

district_summary = pd.DataFrame({
                    "Total Number of Schools": [total_number_schools],
                    "Total Number of Students": [total_number_students],
                    "Total Budget": [total_budget],
                    "Average Math Score": [avg_math_score],
                    "Average Reading Score": [avg_reading_score],
                    "% Passing Math": [perc_pass_math],
                    "% Passing Reading": [perc_pass_reading],
                    "% Overall Passing": [overall_percent]
                                })
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Name

school_group = school_data_complete.set_index("school_name").groupby(["school_name"])
```

```python
#School Type

school_type = school_data.set_index("school_name")["type"]
```

```python
#Total Students

number_students = school_group["Student ID"].count()
```

```python
#Total School Budget

school_budget = school_data.set_index("school_name")["budget"]
```

```python
#Per Student Budget

per_student = school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"]
```

```python
#Average Math Score

avg_math = school_group["math_score"].mean()
```

```python
#Average Reading Score

avg_reading = school_group["reading_score"].mean()
```

```python
#% Passing Math

pass_math = school_data_complete[school_data_complete['math_score']
                                 >= 70].groupby('school_name')['Student ID'].count()/number_students
```

```python
#% Passing Reading

pass_read = school_data_complete[school_data_complete['reading_score']
                                 >= 70].groupby('school_name')['Student ID'].count()/number_students
```

```python
#% Overall Passing

pass_overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] 
                                    >= 70)].groupby('school_name')['Student ID'].count()/number_students
```

```python
#School Summary Table

school_summary = pd.DataFrame({
                                "School Type": school_type,
                                "Total Students": number_students,
                                "Total School Budget": school_budget,
                                "Per Student Budget": per_student,
                                "Average Math Score": avg_math,
                                "Average Reading Score": avg_reading,
                                "% Passing Math": pass_math,
                                "% Passing Reading": pass_read,
                                "% Overall Passing": pass_overall
                                })
```

```python
#School Summary Table Format

school_summary.style.format({
                                "Total Students": '{:,}', 
                                "Total School Budget": "${:,.2f}", 
                                "Per Student Budget": "${:.2f}",
                                "Average Math Score": "{:.6f}", 
                                "Average Reading Score": "{:.6f}", 
                                "% Passing Math": "{:.6%}", 
                                "% Passing Reading": "{:.6%}", 
                                "% Overall Passing": "{:.6%}"
                                })
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Bins to segment and sort values

bins = [0, 585, 629, 644, 6800]
bin_name = ['< $585', "$585 - 630", "$630 - 645", "$645- 680"]
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = bin_name)

#Group by School Spending

by_spending = school_data_complete.groupby('spending_bins')

#Define Averages and Percentages

math_m = by_spending['math_score'].mean()
read_m = by_spending['reading_score'].mean()
math_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
read_pass = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
overall_pass = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
```

```python
#Scores by Spending Summary Table

scores_by_spending = pd.DataFrame({
    "Average Math Score": math_m,
    "Average Reading Score": read_m,
    '% Passing Math': math_pass * 100,
    '% Passing Reading': read_pass * 100,
    "% Overall Passing": overall_pass * 100
            
})
scores_by_spending
```
